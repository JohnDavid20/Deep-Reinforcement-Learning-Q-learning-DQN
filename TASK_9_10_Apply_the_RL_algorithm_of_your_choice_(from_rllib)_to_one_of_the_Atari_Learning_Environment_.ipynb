```python
#I referred to the rllib and gymanesian document to implememt the algorithm and made modifications to suit our environment.https://docs.ray.io/en/latest/rllib/rllib-algorithms.html#dqn

#Import libary for the task 
import gym
from ray import tune
from ray.rllib.algorithms.dqn import DQN
from ray.rllib.utils.spaces.repeated import Repeated
from ray.rllib.algorithms.dqn.dqn import DQNConfig
from ray import air
from ray import tune
import gym
import ray
import ray.rllib.algorithms.dqn as dqn
from ray import tune
import ray
import ray.rllib.algorithms.dqn as dqn
from ray.tune.logger import pretty_print
from ray.rllib.algorithms.dqn.dqn import DQNConfig

# Define the environment name
env_name= gym.make("ALE/SpaceInvaders-v5")
#env_name=gym.make("CartPole-v1")

def evaluation_fn(result):
    return result['episode_reward_mean']

def objective_fn(config, reporter):
    rewards = [] #setting reward list which i will append to get accumalative reward
    trainer = dqn.DQN(config=config, env=env_name)
    for i in range(20):
        result = trainer.train()
        intermediate_score = evaluation_fn(result)
        rewards.append(result['episode_reward_mean'])# appeing the reward to the list
        print(f"episode_reward_max: {result['episode_reward_max']}, mean_reward: {result['episode_reward_mean']}")

#I tune the improvements of DQN and specify the parameters, hyperparameters to be use during the training
config = {
    "env":env_name,
    "double_q": True,
    "dueling": True,
    "Rainbow":True,
    "n_step": 3,
    "target_network_update_freq": 500,
    "capacity": 5000,
    "exploration_final_eps": 0.02,
    "gamma": 0.79,
    "lr": 0.00025,
    "schedule_max_timesteps": 20000,
    "timesteps_per_iteration": 1000,
    "train_batch_size": 32,
    "sample_batch_size": 4,
    "num_workers": 1
}

analysis = tune.run(objective_fn, metric="mean_reward", mode="max", num_samples=1, config=config)
```
